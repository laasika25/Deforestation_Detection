```python
plt.figure(figsize=(8, 6))
sns.countplot(x='type', data=df)
plt.title('Distribution of Fire Types')
plt.xlabel('Fire Type')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['confidence'], bins=20, kde=True)
plt.title('Distribution of Confidence')
plt.xlabel('Confidence')
plt.ylabel('Frequency')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Pie chart of fire types
plt.figure(figsize=(6, 6))
df['type'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=sns.color_palette('Set2'))
plt.title('Fire Type Distribution (Pie Chart)')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='type', y='confidence', data=df, palette='Set3')
plt.title('Confidence by Fire Type (Box Plot)')
plt.xlabel('Fire Type')
plt.ylabel('Confidence')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```
